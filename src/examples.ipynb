# Examples of how to use this repo to simulate static/dynamic E and B fields through visualizations, also incorporating the affects of special relativity by applying a Lorentz transformation for dynamic E fields.

### Simulate a static E field example:

```python
from efield import PointCharge

charge = PointCharge(5, loc = [3,5])
charge.plot(rmax = 0.05, step = 0.0001, figsize = [8,8])
```

### You can also specify a system of charges and plot the corresponding electric field.

```python
from efield import PointChargeSystem 
system_config = {
"q1" : 
    {"charge": 7*10**5, "loc": [1, 1]}, 
"q2": 
    {"charge": 4*10**6, "loc": [1.05, 1.05]},
"q3": 
    {"charge": 2*10**7, "loc": [1.03, 0.9]}
}

system = PointChargeSystem(charge_config = system_config)
system.plot(rmax = 0.05, figsize = [8,8], step = 0.01)
```

### As an alternative to passing in a dictionary, you can create a .json config file to pass in to PointChargeSystem:

```python
from efield import PointChargeSystem 
system = PointChargeSystem()
system.load_charge_config(config_path = '../examples/ex_charge_config.json')
system.plot(rmax = 0.05, figsize = [8,8], step = 0.01)
```
